```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
df = pd.read_csv('energydata_complete.csv')
```

```python
df.head()
```

```python
column_names = {'Appliances': 'appliances','T1':'temp_kitchen', 
                'RH_1': 'rel_humidity_kitchen', 'T2': 'temp_living_room', 'RH_2': 'rel_humidity_living_room',
                'T3': 'temp_laundry_rooom', 'RH_3': 'rel_humidity_laundry_room', 'T4': 'temp_office_room',
                'RH_4': 'rel_humidity_office_room', 'T5':'temp_bathroom', 'RH_5': 'rel_humidity_bathroom', 
                'T6': 'temp_outside_north', 'RH_6': 'rel_humidity_outside_north','T7': 'temp_ironing_rooom',
                'RH_7': 'rel_humidity_ironing_room', 'T8': 'temp_teenager_room_2','RH_8': 'rel_humidity_teenager_room_2',
                'T9': 'temp_parents_room', 'RH_9': 'rel_humidity_parents_room', 'T_out': 'temp_outside', 'Press_mm_hg': 'pressure',
                'RH_out': 'rel_humidity_outside', 'Windspeed': 'wind_speed', 'Visibility': 'visibility', 
               'Tdewpoint': 'temp_dew_point', 'rv1': 'random_variable_1', 'rv2': 'random_variable_2'}
```

```python
df = df.rename(columns=column_names)
```

```python
df.head()
```

```python
df.columns
```

```python
df.head()
```

```python
df = df.drop(columns = ['date', 'lights'], axis = 1)
```

```python
scaler = MinMaxScaler()

normalised_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
features_df = normalised_df.drop(columns=['appliances'], axis = 1)
df_target = normalised_df['appliances']

X_train, X_test, y_train, y_test = train_test_split(features_df, df_target, test_size=.3, random_state=42)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train[['temp_living_room']], X_train.temp_outside)
temp_pred = linear_reg.predict(X_test[['temp_living_room']])
round(r2_score(X_test.temp_outside, temp_pred), 3)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

predicted_values = linear_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 3)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 3) 
```

```python
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)
```

```python
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)
```

```python
def get_weights_df(model, feature, col_name):
    weights = pd.Series(model.coef_, feature.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, X_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, X_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, X_train, 'Lasso_weight')
```

```python
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```

```python
y_pred = linear_model.predict(X_test)
y_pred_ridge = ridge_reg.predict(X_test)
y_pred_lasso = lasso_reg.predict(X_test)
```

```python
round(np.sqrt(mean_squared_error(y_test, y_pred_ridge)), 3)
```

```python
round(np.sqrt(mean_squared_error(y_test, y_pred_lasso)), 3)
```
